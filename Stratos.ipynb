```python
# Assuming all your NVT dataframes are prepared and have 'timestamp' as datetime dtype
dataframes = [stratos_nvt, aioz_nvt, aleph_nvt, arweave_nvt, filecoin_nvt, bitcoin_nvt, ethereum_nvt]

# Rename the 'NVT' column in each dataframe for clarity
for df, name in zip(dataframes, ['stratos', 'aioz', 'aleph', 'arweave', 'filecoin', 'bitcoin', 'ethereum']):
    df.rename(columns={'NVT': f'{name}_NVT'}, inplace=True)

# Merge all dataframes on 'timestamp'
from functools import reduce
merged_df = reduce(lambda left, right: pd.merge(left, right, on='timestamp', how='outer'), dataframes)

# Filter rows where 'timestamp' is from 2021 until now
merged_df = merged_df[merged_df['timestamp'] >= pd.to_datetime('2021-01-01')]
```

```python
# Ensure the 'timestamp' column is sorted
merged_df.sort_values('timestamp', inplace=True)

plt.figure(figsize=(30, 20))  # Adjust size as needed

# Plot each NVT
for name in ['stratos', 'aioz', 'aleph', 'arweave', 'filecoin', 'bitcoin', 'ethereum']:
    plt.plot(merged_df['timestamp'], merged_df[f'{name}_NVT'], label=f'{name} NVT')

plt.title('NVT Ratios Comparison from 2021 Until Now')
plt.xlabel('Date')
plt.ylabel('NVT Ratio')
plt.legend()  # Show legend to identify each line
plt.xticks(rotation=45)  # Rotate date labels for better readability

plt.tight_layout()
plt.show()
```

```python
# Ensure the 'timestamp' column is sorted, and filter for dates from 2021 onwards
merged_df = merged_df[merged_df['timestamp'] >= pd.to_datetime('2021-01-01')].sort_values('timestamp')

plt.figure(figsize=(15, 10))  # Adjust size as needed

# Plot each NVT
for name in ['stratos', 'aioz', 'aleph', 'arweave', 'filecoin', 'bitcoin', 'ethereum']:
    plt.plot(merged_df['timestamp'], merged_df[f'{name}_NVT'], label=f'{name} NVT', marker='o')

plt.title('NVT Ratios Comparison from 2021 Until Now')
plt.xlabel('Date')
plt.ylabel('NVT Ratio')
plt.legend()  # Show legend to identify each line
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.grid(True)  # Add grid for better readability of values

plt.tight_layout()
plt.show()
```

```python

# Create a date range that covers all dates from the start of 2023 to the last date in your data
date_range = pd.date_range(start='2023-01-01', end=merged_df['timestamp'].max())

# Reindex each series in the dataframe to this complete date range, filling missing dates with NaN
# Then, interpolate the NaN values to make the lines continuous
for name in ['stratos_NVT', 'aioz_NVT', 'aleph_NVT', 'arweave_NVT', 'filecoin_NVT', 'bitcoin_NVT', 'ethereum_NVT']:
    series = merged_df.set_index('timestamp')[name].reindex(date_range).interpolate(method='linear')
    plt.plot(series.index, series, label=name, marker='o')

plt.title('Continuous NVT Ratios Comparison from 2023')
plt.xlabel('Date')
plt.ylabel('NVT Ratio')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
